Using the ConstructionTimeSeriesDataV2.csv, these are the graphics which relate below. Within this document I will compare Private Construction against Total Construction prices.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

ConstructionTS = pd.read_csv("ConstructionTimeSeriesDataV2.csv")
# ConstructionTS
```

```python
x = ConstructionTS.index
y = ConstructionTS['Total Construction']
plt.figure(figsize = (24,4))
plt.plot(x,y,label = 'Total Construction')
plt.xlabel('Month', fontsize = 14)
plt.ylabel('Spending', fontsize = 14)
plt.axis([0, max(ConstructionTS['Month'])+1, 0 , max(ConstructionTS['Total Construction'])*1.05], fontsize = 14)
plt.show()
```

In the graphic above, we can see the raw data of the Total Constuction spending which has occurred in our data set.

```python
plt.scatter(ConstructionTS[['Private Construction']], ConstructionTS[['Total Construction']], alpha=0.5)
x = np.array(ConstructionTS['Private Construction'])
y = np.array(ConstructionTS['Total Construction'])

plt.suptitle('Private vs. Total Construction')
plt.xlabel("Private Construction")
plt.ylabel("Total Construction")

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--")
plt.show()
```

The graph above shows the relationship in costs with Private and Total Constuction spending. As expected they have a linear positive relationship.

```python
x = ConstructionTS.Month
y = ConstructionTS['Total Construction']
y1 = ConstructionTS['Private Construction']
plt.plot(x, y, label = 'Total Construction')
plt.plot(x, y1, label = 'Private Construction')
plt.xlabel('Month')
plt.ylabel('Construction Spending')
plt.axis([x.min(), x.max(), 0, 1.1*y.max()])
plt.legend()
plt.savefig('initialGraph.jpg')
```

Here, we can see the cyclical patterns which exists in spending for both Total and Private Construction.

```python
x = ConstructionTS.Month[0:12]
y1 = ConstructionTS['Total Construction'][0:12]
y2 = ConstructionTS['Private Construction'][0:12]

x_labs = ['', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

fig, ax = plt.subplots()

ax.plot(x, y1, label = 'Total Construction')
ax.plot(x, y2, label = 'Private Construction')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(loc = 'lower center', prop = {'family':'Times New Roman', 'size':'large'})
ax.set_xlim(0, x.max()+1)
ax.set_ylim(0, 1.1*y1.max())

ax.xaxis.set_label_text('Month', fontsize = 18, fontname = 'Times New Roman')
ax.xaxis.set_ticks(range(0,13))
ax.xaxis.set_ticklabels(x_labs)
ax.xaxis.set_tick_params(which = 'both', top = 0, bottom = 1, labelbottom = 1)

ax.yaxis.set_label_text('Spending', fontsize = 18, fontname = 'Times New Roman')
ax.yaxis.set_tick_params(which = 'both', right = 0, left = 1, labelleft = 1)

fig.set_size_inches(7, 5)
fig.suptitle('Private vs. Total Construction', fontsize = 'xx-large', fontname = 'Times New Roman')
fig.tight_layout()
fig.savefig('Total vs. Private Construction Month.jpg')
```

In the above graph, we can see the monthly spending for Total and Private. Like past graphs, we can see the archs match eachother pretty closely. In the graph below, we can see each spending broken down.

```python
x = ConstructionTS.Month[0:12]
y1 = ConstructionTS['Total Construction'][0:12]
y2 = ConstructionTS['Private Construction'][0:12]
x_labs = ['', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

fig, (ax1, ax2) = plt.subplots(1,2,sharey = True)

ax1.plot(x, y1, label = 'Total Construction')
ax2.plot(x, y2, label = 'Private Construction')

for ax in fig.axes:
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.legend(loc = 'upper right', prop = {'family': 'Times New Roman', 'size':'large'})
    ax.set_xlim(0, x.max()+1)
    ax.set_ylim(0, 1.1*y1.max())
    ax.xaxis.set_label_text('Month', fontsize = 18)
    ax.xaxis.set_ticks(range(0,13))
    ax.xaxis.set_ticklabels(x_labs)
    ax.xaxis.set_tick_params(which = 'both', right = 0, bottom = 1, labelbottom = 1)
    ax.yaxis.set_label_text('Spending', fontsize = 18, fontname = 'Times New Roman')
    ax.yaxis.set_tick_params(which = 'both', right = 0, left = 1, labelleft = 1)
    
fig.set_size_inches(12, 5)
fig.tight_layout()
plt.savefig('two for one.jpg')
```
